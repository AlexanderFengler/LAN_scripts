```python
import os
import time
import pickle

dir_name = 'data/lan_mlp/training_data_0_nbins_0_n_200000/ds_conflict_drift/'
list_of_files = filter(lambda x: os.path.isfile(os.path.join(dir_name, x)),
                       os.listdir(dir_name))

list_of_files = sorted( list_of_files,
                        key = lambda x: os.path.getmtime(os.path.join(dir_name, x))
                      )
```

```python
for i in range(len(list_of_files)):
    y = pickle.load(open(dir_name + list_of_files[i], 'rb'))
    
    if 'theta' in y['model_config']['params']:
        y['model_config']['name'] = 'ds_conflict_drift_angle'
        new_file_name = list_of_files[i][:(25 + 7)] + 'angle_' + list_of_files[i][(25 + 7):]
        pickle.dump(y, open(dir_name[:-1] + '_angle/' + new_file_name, 'wb'))
        print(i)
    else:
        print('passed', i)
        pass
```
