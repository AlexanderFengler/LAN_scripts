```python
dir_name = 'data/torch_models/'
```

```python
list_of_files = filter(lambda x: os.path.isfile(os.path.join(dir_name, x)),
                       os.listdir(dir_name))

list_of_files = sorted( list_of_files,
                        key = lambda x: os.path.getmtime(os.path.join(dir_name, x))
                      )
```

```python
tm_mon_constraint = 12 # NOTE: 12 means don't delete based on month
tm_year_constraint = 2022 # NOTE: below 2022 will be deleted
rm_ = 0

for file_name in list_of_files:
    file_path = os.path.join(dir_name, file_name)
    timestamp_str = time.strftime(  '%m/%d/%Y :: %H:%M:%S',
                                    time.gmtime(os.path.getmtime(file_path)))
    tm_year = time.gmtime(os.path.getmtime(file_path)).tm_year
    tm_mon = time.gmtime(os.path.getmtime(file_path)).tm_mon
    print(timestamp_str, ' -->', file_name, ' FOUND')
    if (tm_year < tm_year_constraint) or ((tm_year > tm_year_constraint) and (tm_mon < tm_mon_constraint))
        if rm_: 
            os.remove(file_path)
            print(timestamp_str, ' -->', file_name, ' DELETED')
    
```
